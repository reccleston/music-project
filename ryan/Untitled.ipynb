```python
import pandas as pd
```

```python
data = pd.read_csv('../data/top10s.csv', encoding='cp1252').drop(columns=['Unnamed: 0'])
```

```python
data.head()
```

```python
correaltion_vals = data.corr().drop(columns={'year'})[1:]
```

```python
correaltion_vals
```

```python
row_num = len(correaltion_vals.columns)
cols = correaltion_vals.columns

vals = {
    'feat1': [],
    'feat2': [],
    'val':[]
}

for i in range(0, row_num):
    for ii in range(0,row_num):
        vals['feat1'].append(cols[i])
        vals['feat2'].append(cols[ii])
        vals['val'].append(correaltion_vals.iloc[i, ii])

pd.DataFrame(vals).to_csv('data/corr_heatmap_vals.csv', index=False)
```
